### Quest√£o 2

```
import numpy as np
```

```
#Fun√ß√£o Fatorial:
def fatorial(x):
    if x == 0 or x == 1:
        return 1 
    else:
        return x * fatorial(x - 1) 
      
```

```
#Montando o vetor:
for i in range(0,9):
  if i%2 == 0 or i == 0:
    vetor[i] = (3 ** i) + (7 *(fatorial(i)))
  else:
    vetor[i] = (2 ** i) + 4 * (np.log(i))
print(vetor)
```

```
#Encontrando a posi√ß√£o:
X = vetor.index(max(vetor)) + 1
X
```

```
#Encontrando a media:
media = np.average(vetor, axis=0)
Y = media.round(2)
Y
```

```
print("A posi√ß√£o do maior elemento, de 1 a 10 √©:", X)
print("A m√©dia dos elementos do vetor √©:", Y)
```
